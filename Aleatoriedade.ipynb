```python
# importando as bibliotecas inicias
import pandas as pd
import numpy as np
import random

random.seed(1522)
```

```python
# instanciando uma variável com 2000 elementos pseudo-aleatórios
num_elements = 2000

# instanciando uma lista vazia para guardar os elementos gerados
list_numbers = []

for i in range(0, num_elements):
    # gerando números pseudo-aleatórios entre -1 e 1
    list_numbers.append(random.choice([-1, 1]))
```

```python
# com os números gerados, criando um data frame
df = pd.DataFrame(
    data=list_numbers, columns=['number']
)

# visualizando o data frame
df
```

```python
# criando uma coluna no df ('sum_') para o método ".cumsum()" para realizar o cálculo acumulado dos números aleatórios
df['sum_'] = df['number'].cumsum()
```

```python
# com a função ".rolling()" iremos criar as médias móvies de 20 e 200, e para posteriormente realizarmos a cálculo da média
df['mean_20'] = df['sum_'].rolling(20).mean()
df['mean_200'] = df['sum_'].rolling(200).mean()
```

```python
# visualizando o data frame final
df
```
